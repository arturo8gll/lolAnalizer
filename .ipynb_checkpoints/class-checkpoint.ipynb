# Variable discretas y continuas

```python
cont=list(df.select_dtypes(include=['float','int']).columns)
```

```python
disc=list(df.select_dtypes(include=['object']))
```

```python
def missingsPercentages(df):
  percentageOfMiss=[]
  for i in df.columns.values:
    miss=[i,df[i].isnull().sum()/df.shape[0]*100]
    percentageOfMiss.append(miss)
  return pd.DataFrame(percentageOfMiss,columns=['feat','missings%']).sort_values('missings%',ascending=0).reset_index(drop=True)
```

```python
missings=missingsPercentages(df[cont])
```

```python
missings
```

```python
l_miss=missings[missings["missings%"]>0]["feat"].values
```

```python
l_miss
```

```python
from sklearn.impute import SimpleImputer
im = SimpleImputer(strategy='median')
```

```python
im.fit(df[l_miss])
```

```python
imputedTable=df.copy()
imputedTable[l_miss]=im.transform(imputedTable[l_miss])
```

```python
missingsPercentages(imputedTable).head(50)
```
