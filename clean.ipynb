```python
import pandas as pd
import os
from tqdm import tqdm
```

```python
games=os.listdir("./games/")
```

```python
final=pd.DataFrame()
for i in tqdm(range(len(games))):
    df=pd.read_json(f"./games/{games[0]}")
    for i in range(len(df.participants.values)):
        if df.participants.values[i]==None:
            continue
        players=pd.json_normalize(df.participantIdentities.values[i])    
        game=pd.json_normalize(df.participants.values[i])
        final=final.append(players.merge(game,on="participantId"))
```

```python
final.shape
```

```python
final.to_csv("./games.csv",index=False)
```
